```python
# clean up data
# Only want USA Races, 50k or 50Mi, 2020

# Step 1: Show 50Mi or 50k
#50k
```

```python
df[df["Event distance/length"] == "50km"]
```

```python
#combine 50k and 50mi with isin

df[df["Event distance/length"].isin(["50km", "50mi"])]
```

```python
df[(df["Event distance/length"].isin(["50km", "50mi"])) & (df["Year of event"] == 2020)]
```

```python
df[df["Event name"].str.contains("USA")]
```

```python
# Combine all the filters

df2 = df[(df["Event distance/length"].isin(["50km", "50mi"])) & (df["Year of event"] == 2020) & (df["Event name"].str.contains("USA"))].copy()
```

```python
df2.head(10)
```

```python
#remove (USA) from the event name

df2["Event name"] = df2["Event name"].str.replace(" (USA)", "")
```

```python
df2.head(10)
```

```python
# Clean up the age

df2["athlete_age"] = 2020 - df2["Athlete year of birth"]
```

```python
# Remove the h from athlete performance

df2["Athlete performance"] = df2["Athlete performance"].str.replace("h", "")
```

```python
# drop the columns we don't need (Athlete Club, Athlete country, Athlete year of birth, Athlete age category)

df2.drop(columns=["Athlete club", "Athlete country", "Athlete year of birth", "Athlete age category"], inplace=True)
```

```python
df2.head()
```

```python
#clean up null values

df2.isna().sum()
```

```python
df2[df2["athlete_age"].isna() == 1]
```

```python
df2.dropna(subset=["athlete_age"], inplace=True)
```

```python
df2.isna().sum()
```

```python
#check for dupes

df2.duplicated().sum()
```

```python
#fix types

df2.dtypes
```

```python
df2["athlete_age"] = df2["athlete_age"].astype(int)
```

```python
df2["Athlete average speed"] = df2["Athlete average speed"].astype(float)
```

```python
df2.dtypes
```

```python
#rename columns

# Year of event                  int64
# Event dates                   object
# Event name                    object
# Event distance/length         object
# Event number of finishers      int64
# Athlete performance           object
# Athlete gender                object
# Athlete average speed        float64
# Athlete ID                     int64

df2.rename(columns={"Year of event": "year", "Event dates": "date", "Event name": "name", "Event distance/length": "distance", "Event number of finishers": "finishers", "Athlete performance": "performance", "Athlete gender": "gender", "Athlete average speed": "avg_speed", "Athlete ID": "id"}, inplace=True)
```

```python
df2.head()
```

```python
# find the fastest runner in the dataset

fastest_runner = df2[df2["avg_speed"] == df2["avg_speed"].max()]
```

```python
# find the slowest runner in the dataset

slowest_runner = df2[df2["avg_speed"] == df2["avg_speed"].min()]
```

```python
df_fastest_slowest = pd.concat([fastest_runner, slowest_runner])
df_fastest_slowest
```

```python
#charts and graphs
import matplotlib.pyplot as plt

sns.histplot(df2["athlete_age"])


plt.ylabel("Number of Athletes")
plt.xlabel("Age")
plt.show()
```

```python
sns.displot(df2[df2["distance"] == "50km"]["avg_speed"], kde=True)

plt.ylabel("Number of Athletes")
plt.xlabel("Average Speed")
plt.show()
```

```python
sns.violinplot(data=df2, x="distance", hue="gender", y="avg_speed", split=True, inner="quartile", linewidth=1)

plt.xlabel("Distance")
plt.ylabel("Average Speed")
plt.show()
```

```python
# Difference in speed between male and female in 50km and 50mi races

df2.groupby(["distance","gender"])["avg_speed"].mean()
```

```python
# What age group are the best in the 50m Race (20 + races minimum)

df2.query("distance == '50mi'").groupby("athlete_age")["avg_speed"].agg(["mean", "count"]).sort_values("mean", ascending=False).query("count >= 20").head(10)
```

```python
# What ge group are the worst in the 50m Race (20 + races minimum)

df2.query("distance == '50mi'").groupby("athlete_age")["avg_speed"].agg(["mean", "count"]).sort_values("mean", ascending=True).query("count >= 20").head(10)
```

```python
# The runners are slower in summer or winter?

# Spring = 3 - 5
# Summer = 6 - 8
# Fall = 9 - 11
# Winter = 12 - 2

df2["date"] = df2["date"].str.replace(".", "/")

df2["month"] = df2["date"].str.split("/").str[1].astype(int)

df2["season"] = df2["month"].apply(lambda x: "Winter" if x > 11 else "Fall" if x > 8 else "Summer" if x > 5 else "Spring")

avg_speed_seasons = df2.groupby("season")["avg_speed"].agg(["mean", "count"]).sort_values("mean", ascending=False)

avg_speed_seasons
```

```python
sns.barplot(data=avg_speed_seasons, x=avg_speed_seasons.index, y="mean", palette="Paired", hue="count", legend=False, width=0.5)

plt.ylabel("Average Speed")
plt.xlabel("Season")
plt.ylim(6, 8)
plt.show()
```
